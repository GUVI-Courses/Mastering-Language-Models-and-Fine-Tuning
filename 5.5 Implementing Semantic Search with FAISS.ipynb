```python
!pip install faiss-cpu # Install faiss-cpu for CPU usage. If you have a GPU, you may consider 'faiss-gpu' instead.
import numpy as np
import faiss

# Create a sample dataset
data = np.random.rand(100, 128).astype('float32')

# Create a FAISS index
index = faiss.IndexFlatL2(128)

# Add the data to the index
index.add(data)

# Define a query vector
query = np.random.rand(1, 128).astype('float32')

# Search for similar vectors
D, I = index.search(query, k=10)

# Print the results
print("Similar vectors:")
print(D)
print("Indices of similar vectors:")
print(I)
```
